## 2. Vizualizace dat

 ### 2.0. Import a načtení dat

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('test.csv')
```

### 2.1. Věk pasažérů letadel

```python
x = data['Age'].value_counts().sort_index()
plt.plot(x, '-', linewidth=2)
plt.xlabel('Věk')
plt.ylabel('Výskyt')
plt.title('Věk')
plt.show()
```

### 2.2. Typ cesty

```python
x = data['Type of Travel'].value_counts()
plt.pie(x.values, labels=x.index)
plt.title('Typ cesty')
plt.show()
```

### 2.3. Vzdálenost letu

```python
x = data['Flight Distance'].value_counts().sort_index()
plt.bar(x.index, x.values, width=50)
plt.xlabel('Vzdálenost (km)')
plt.ylabel('Výskyt')
plt.title('Vzdálenost')
plt.ylim(min(x.values), max(x.values))
plt.xlim(0, 4000)
plt.show()
```

### 2.4. Průměrné zpoždění odletu

```python
x = data['Departure Delay in Minutes'].value_counts().sort_index()
plt.plot(x, '-', linewidth=1)
plt.xlabel('Průměrné zpoždění odletu (min)')
plt.ylabel('Výskyt')
plt.title('Zpoždění')
plt.show()
```

### 2.5. Čistota letadla

```python
x = data['Cleanliness'].value_counts().sort_index()
plt.bar(x.index, x.values, linewidth=2)
plt.xlabel('Hodnocení (5 nejlepší)')
plt.ylabel('Výskyt')
plt.title('Čistota')
plt.show()
```
